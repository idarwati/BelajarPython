```python
import numpy as np # karna akan membuat DataFrame baru
```

```python
np.random.seed(100)
```

```python
angka = []
for i in range(0,100):
  angka.append(i)
```

```python
angka
```

```python
baru = pd.DataFrame(np.random.rand(100,5),angka, columns=['A','B','C','D','E']) #tulisan columns bisa tidak ditulis, jadi langsung masukkan tanda [], hasilnya akan sama.
```

```python
baru.head()
```

```python
baru.plot.area(alpha=0.4)
```

```python
baru2 = pd.DataFrame(np.random.rand(4,5),[0,1,2,3],columns=['A','B','C','D','E'])
```

```python
baru2
```

```python
baru2.plot.bar()
```

```python
baru2.plot.bar(stacked=True)
```

```python
baru2.C.plot.bar()
```

```python
baru2['C'].plot.bar()
```

```python
baru.plot.scatter(x='A', y='B', c='E') #c adalah colour, warna yang ada dibawah itu sesuai nilai yang ada di kolom E
```

```python
baru.plot.scatter(x='A', y='B', c='E', cmap='ocean') #cmap digunakan untuk mengubah warna
```

```python
baru.plot.scatter(x='A', y='B', s=baru.D)
```

```python
baru.plot.scatter(x='A', y='B', s=baru.D*100) #skalanya berdasarkan D
```

```python
baru.plot.scatter(x='A', y='B', s=baru.D*100, c='C') #skalanya berdasarkan D, warna berdasarkan C
```

```python
baru.plot.scatter(x='A', y='B', s=baru.D*100, c='C', cmap='Blues') #skalanya berdasarkan D, warna berdasarkan C
```

```python
baru.plot.box()
```

```python
baru.plot.line(y='B')
```

```python
baru.plot.line(y='B', lw=5) #lw untuk mengatur ketebalan line
```

```python
baru.plot.line(y='B', lw=0.5, figsize=(15,3)) #lw untuk mengatur ketebalan line, figsize untuk mengatur panjang dan lebar grafik
```

```python
baru.plot.hexbin(x='A', y='B')
```

```python
baru.plot.hexbin(x='A', y='B', gridsize=25) #gridsize digunakan untuk mengatur ketebalan hexbin
```
