```python
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Grafik Kuadrat")
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('Sumbu X1')
plt.ylabel('Sumbu Y1')
plt.title('Grafik Kuadrat 1')

plt.subplot(122)
plt.plot(y,x,color='blue')
plt.xlabel('Sumbu X2')
plt.ylabel('Sumbu Y2')
plt.title('Grafik Kuadrat 2')
plt.tight_layout()
plt.show()
```

# **Matplotlib dengan object oriented**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plt.plot() <- cara konvensional, fungsional
```

```python
x = np.arange(1,20,0.5) # dimulai dari angka 1 hingga 20, tapi berhenti sebelum angka 20 dengan jarak masing2 0.5
y = x**2
```

```python
x
```

```python
y
```

```python
# versi Matplotlib OOP

# versi 1
```

```python
fig = plt.figure() #menyiapkan sebuah kanvas
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # urutan penulisan ukurannya left, bottom, width, height
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Grafik Kuadrat')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.8, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x,y, color='red')
axes1.set_xlabel('Sumbu X besar')
axes1.set_ylabel('Sumbu Y besar')
axes1.set_title('Grafik Kuadrat besar')

axes2.plot(y,x)
axes2.set_xlabel('Sumbu X kecil')
axes2.set_ylabel('Sumbu Y kecil')
axes2.set_title('Grafik Kuadrat kecil')

plt.show()
```

```python
# Versi 2
```

```python
fig, axes = plt.subplots(2,1) #(2,1) artinya index/baris=2 dan kolom=1
```

```python
fig, axes = plt.subplots(2,1)
fig.tight_layout() #python mencari jarak yang pas
```

```python
fig, axes = plt.subplots(3,2)
fig.tight_layout() #python mencari jarak yang pas
```

```python
fig, axes = plt.subplots(2,1)

# Plot for first axes
axes[0].plot(x,y, color='red')
axes[0].set_title('Grafik Kuadrat 1')
axes[0].set_xlabel('Sumbu X1')
axes[0].set_ylabel('Sumbu Y1')

# Plot for second axes
axes[1].plot(x,y)
axes[1].set_title('Grafik Kuadrat 2')
axes[1].set_xlabel('Sumbu X2')
axes[1].set_ylabel('Sumbu Y2')

fig.tight_layout() #python mencari jarak yang pas
```

```python
fig, axes = plt.subplots(1,2)

# Plot for first axes
axes[0].plot(x,y, color='red')
axes[0].set_title('Grafik Kuadrat 1')
axes[0].set_xlabel('Sumbu X1')
axes[0].set_ylabel('Sumbu Y1')

# Plot for second axes
axes[1].plot(y,x,color='green')
axes[1].set_title('Grafik Kuadrat 2')
axes[1].set_xlabel('Sumbu X2')
axes[1].set_ylabel('Sumbu Y2')

fig.tight_layout() #python mencari jarak yang pas
```

# **Matplotlib** (pengaturan grafik)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
```

```python
fig = plt.figure(figsize=(4,4), dpi=200) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(4,4), dpi=100) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(8,4), dpi=100) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(1,4), dpi=100) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,8), dpi=100)

axes[0].plot(x,y)
axes[0].set_title('Grafik 1')

axes[1].plot(x,y)
axes[1].set_title('Grafik 2')

plt.tight_layout()

plt.show()
```
