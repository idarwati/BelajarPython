# Hukum Eksponen

```python
import sympy as sym
from IPython.display import display, Math
```

```python
display(Math('x^{x} \\cdot x^{z} = x^{y+z}'))
```

```python
display(Math('\\frac {x^{y}}{x^{z}} = x^{y} x^{-z} = x^{y-z}'))
```

```python
display(Math('x^{y} y^{z} = x^{y} y^{z}'))
```

```python
x, y, z = sym.symbols('x y z')
```

```python
p1 = (x**y) * (x**z)
```

```python
display(p1)
```

```python
# simplify (menyederhanakan)

display(sym.simplify(p1))
```

```python
p2 = x**y / x**z
```

```python
display(p2)
display(Math(sym.latex(p2)))
```

```python
p3 = x**y * y**z
display(p3)
display(sym.simplify(p3))
```

# Hukum Asosiatif, Komutatif & Distributif

```python
import sympy as sym
from IPython.display import display, Math
```

```python
x, y, z = sym.symbols('x, y, z')
```

```python
# Asosiatif

p1 = 5*(x*y)
p2 = (5*x)*y
```

```python
p1 - p2
```

```python
p1 == p2
```

```python
display(p1)
```

```python
display(p2)
```

```python
print(p1.subs({x:5, y:8}))
```

```python
print(p2.subs({x:5, y:8}))
```

```python
# Komutatif

p3 = 5*x*y
p4 = x*y*5
p5 = y*5*x
```

```python
p3 == p4 == p5
```

```python
print(p3.subs({x:5, y:8}))
print(p4.subs({x:5, y:8}))
print(p5.subs({x:5, y:8}))
```
