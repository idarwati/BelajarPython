```python
x = np.linspace(-2, 5, 200) #differences itu aslinya lebih kecil (dikurangi 1), klo 200 jadi aslinya 199
f = x**2 + 1
dx = np.diff(x)
df = np.diff(f)
```

```python
plt.plot(x,f)
```

```python
plt.plot(x[:-1],df/dx)
```

```python
plt.plot(x,f, label='f')
plt.plot(x[:-1], df/dx, label='df/dx')
plt.grid()
plt.legend()

plt.show()
```

```python
# slope nya adalah 2x
# slope x = 2 adalah 4

# f(x) = x**2 + 1
# jika x = 3, maka f = 10
# jika x = 1, maka f = 2
# slope = (perubahan f)/(perubahan x) = (10-2)/(3-1) = 8/2 = 4

# klo titik di atas 0 berarti mencari nilai maksimum, klo dibawah 0 berarti mencari nilai minimum
```

```python
plt.plot(x,f, label='f')
plt.plot(x[:-1], df/dx, label='df/dx')
plt.grid()
plt.legend()
plt.axis([-2,5,-10,20])

plt.show()
```
