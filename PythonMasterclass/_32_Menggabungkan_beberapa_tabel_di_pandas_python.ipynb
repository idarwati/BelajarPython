```python
satu = pd.DataFrame({'A':['a0','a1','a2','a3','a4','a5'],
                     'B':['b0','b1','b2','b3','b4','b5'],
                     'C':['c0','c1','c2','c3','c4','c5'],
                     'D':['d0','d1','d2','d3','d4','d5']},
                    index=[0,1,2,3,4,5])
```

```python
bisnis = pd.DataFrame({'Perusahaan':['Telkom', 'Telkom', 'Pertamina', 'BRI', 'Pertamina', 'KFC', 'KFC', 'Indosat'],
                       'Karyawan':['Joko', 'Budi', 'Susi', 'Iriana', 'Bambang', 'Udin', 'Bahlil', 'Intan'],
                       'Usia':[30,40,55,35,46,53,20,19]})
```

```python
bisnis
```

```python
bisnis.groupby('Perusahaan')
```

```python
bisnis.groupby('Perusahaan').sum()
```

```python
bisnis.groupby('Perusahaan').count()
```

```python
# Calculate the mean only for the 'Usia' column
bisnis.groupby('Perusahaan')['Usia'].mean()
```

```python
bisnis.groupby('Perusahaan').min()
```

```python
bisnis.groupby('Perusahaan').max()
```

```python
bisnis.groupby('Perusahaan').Usia.min()
```

```python
bisnis.groupby('Perusahaan').describe()
```

```python
bisnis.groupby('Perusahaan').describe().transpose()
```

```python
import pandas as pd
```

```python
satu = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                     'one':[1,2,3,4,5],
                     'two':[6,7,8,9,10]})
```

```python
dua = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                     'three':[1,2,3,4,5],
                     'four':[6,7,8,9,10]})
```

```python
satu
```

```python
dua
```

```python
#Merge
```

```python
baru = pd.merge(satu,dua,how='inner',on='key') #on merupakan label/list, utk belajar lebih lanjut tentang inner bisa membuka link diffen.com
```

```python
baru
```

```python
satu = pd.DataFrame({'key1': ['K0','K0','K1','K2'],
                     'key2': ['K0','K1','K0','K1'],
                     'A': ['A0','A1','A2','A3'],
                     'B': ['B0','B1','B2','B3']})

dua = pd.DataFrame({'key1': ['K0','K1','K1','K2'],
                    'key2': ['K0','K0','K0','K0'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']})
```

```python
satu
```

```python
dua
```

```python
baru2 = pd.merge(satu, dua, how='left', on=['key1','key2'])
```

```python
baru2
```

```python
baru2 = pd.merge(satu, dua, how='right', on=['key1','key2'])
```

```python
baru2
```

```python
baru2 = pd.merge(satu, dua, how='outer', on=['key1','key2'])
```

```python
baru2
```
