What is Linear Algebra?
* Matematikanya data
* Cara data berkomunikasi adalah dengan aljabar linear
  * Matriks
  * Vektor

Vektor
* Vektor = Kumpulan angka
* Vektor secara aljabar (algebraic vectors) ada 2 jenis:
  * Baris
  * Kolom
* Vektor Baris (row vectors) :
  [ 2 9 0 8 12 ]
  [ -21 -3 -54 8]
* Vektor Kolom (colomn vectors):
  [ 1
    2
    3
    4 ]
* Vektor Geometrik (Geometric vectors)
  * 2 dimensi : [ 1 2 ] artinya dari sumbu 0 bergerak 1 ke sumbu x, dan bergerak 2 ke sumbu y.
  * 3 dimensi : [ 1 2 3 ] artinya bergerak 1 ke sumbu x, bergerak 2 ke sumbu y dan bergerak 3 ke sumbu z.

# Practice Aljabar Linear

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math
from IPython import get_ipython
```

```python
baris = np.array([2,3])
kolom = np.array([[1],[4]])
```

```python
print(baris)
```

```python
print(kolom)
```

```python
display(Math(sym.latex(baris)))
```

```python
display(Math(sym.latex(sym.simplify(baris))))
```

```python
display(Math(sym.latex(sym.simplify(kolom))))
```

```python
plt.plot([0, baris[0]], [0, baris[1]])
```

```python
plt.plot([0, baris[0]], [0, baris[1]], 'r', label='Vektor baris')
# Access the scalar values within the nested arrays of the 'kolom' vector
plt.plot([0, kolom[0][0]], [0, kolom[1][0]], 'b', label='Vektor kolom')

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
from IPython import get_ipython
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np


# Definisikan dua vektor yang berbeda untuk diplot
a = np.array([2,0,-3]) # Vektor untuk plot merah
b = np.array([-2,2,4]) # Vektor untuk plot biru (ganti nilai sesuai kebutuhan Anda)

fig = plt.figure(figsize = plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

# Plot vektor 'a' dengan warna merah
ax.plot([0, a[0]], [0, a[1]], [0, a[2]], 'r', linewidth=3, label='Vektor Merah')

# Plot vektor 'b' dengan warna biru
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'b', linewidth=3, label='Vektor Biru')


ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)

# Tambahkan label untuk sumbu
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_zlabel('Sumbu Z')

# Tambahkan legenda untuk membedakan vektor
ax.legend()

plt.show()
```
