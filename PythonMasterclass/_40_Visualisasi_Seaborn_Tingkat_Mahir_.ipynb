```python
sns.pairplot(berlian)
```

```python
sns.PairGrid(berlian)
```

```python
p = sns.PairGrid(berlian)
p.map(plt.scatter) #menggunakan matplotlib
plt.show
```

```python
p = sns.PairGrid(berlian)
p.map_diag(sns.distplot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
```

```python
tabrak = sns.load_dataset('car_crashes')
```

# **FacetGrid**

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
p = sns.FacetGrid(tips, row='smoker', col='day') # output hanya 4 kolom dan baris hanya 2
p.map(sns.distplot, 'total_bill') #kombinasi dari 3 parameter yaitu smoker, day dan total_bill
plt.show()
```

```python
p = sns.FacetGrid(tips, row='smoker', col='time')
p.map(sns.distplot, 'total_bill')
plt.show()
```

```python
p = sns.FacetGrid(tips, row='smoker', col='day') # output hanya 4 kolom dan baris hanya 2
p.map(plt.scatter, 'total_bill', 'tip') #kombinasi dari 3 parameter yaitu smoker, day dan total_bill
plt.show()
```
