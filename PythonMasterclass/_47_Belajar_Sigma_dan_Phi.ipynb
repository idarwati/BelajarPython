# Bilangan
* Penjumlahan bilangan (Summation) : Sigma
* Perkalian bilangan (Products) : Phi (seperti dalam trigonometri)
* Cumulative Sum
* Cumulative Products



```python
b1 = [1,2,3,4,5]
```

```python
sum(b1)
```

```python
import numpy as np
```

```python
np.sum(b1)
```

```python
# products
np.prod(b1)
```

```python
# cumulative summation
np.cumsum(b1)
'''
x1 = 1,
x2 = 1+2,
x3 = 1+2+3,
x4 = 1+2+3+4,
x5 = 1+2+3+4+5
'''
```

```python
b2 = np.array([[1,2,3],[4,5,6]])
```

```python
print(b2)
```

```python
from IPython.display import display, Math, Latex
```

```python
import sympy as sym
```

```python
display(Math(sym.latex(sym.sympify(b2))))
```

```python
np.cumsum(b2)
'''
x1 = 1
x2 = 1+2
x3 = 3+3
x4 = 6+4
x5 = 10+5
x6 = 15+6
'''
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.plot(b1, 'rs-') #'rs-' : grafiknya warna merah, dimana setiap titiknya digabungkan
plt.plot(np.cumsum(b1), 'bo-') #'bo-' : grafiknya warna biru
plt.legend(['b1','kumulatif'])
plt.show()
```

```python
plt.plot(b1, 'rs-')
plt.plot(np.cumprod(b1), 'bo-')
plt.legend(['b1','kum_prod'])
plt.show()
```

```python
bil2 = np.arange(1,5)
```

```python
p1 = np.sum(bil2)
p2 = np.sum(bil2**2)
```

```python
j1 = p1/p2
j2 = 1/p1
```

```python
print(f'Jawaban 1 hasilnya adalah {j1} \nJawaban 2 hasilnya adalah {j2}')
```

```python
j1 == j2
```
