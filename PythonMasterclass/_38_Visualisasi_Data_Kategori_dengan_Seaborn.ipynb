```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
file_id = '1kQHNP1cpFw-44yTaEcbLrvBKY_CtUDwj'
url = f'https://drive.google.com/uc?id={file_id}'
```

```python
dataku = pd.read_csv(url)
```

```python
dataku.head()
```

```python
dataku.describe(include='O') # O untuk data2 ordinal saja
```

```python
# Barplot (histogram)
sns.barplot(x='sex', y='fare', palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
# Barplot (histogram)
sns.barplot(x='fare', y='sex', palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
sns.barplot(x='sex', y='fare', data=dataku, palette={'male':'skyblue', 'female':'pink'}, estimator=sum)
```

```python
sns.barplot(x='sex', y='fare', data=dataku, palette={'male':'skyblue', 'female':'pink'}, estimator=np.std)
```

```python
# countplot
sns.countplot(x='sex', palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
dataku
```

```python
sns.countplot(x='deck', hue='sex', palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
# cara membuat grafiknya berurutan

sns.countplot(x='deck', hue='sex', order=['A','B','C','D','E','F','G'], palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
# Boxplot (box % whishers plot)

sns.boxplot(x='sex', y='fare', palette={'male':'skyblue', 'female':'pink'}, data=dataku)
#semua titik2 diluar garis atas dan bawah dianggap sebagai outlier
# Q3-Q1 = IQR/Inter Quartilee Range
# outlier ataas = Q3 + 1.5 * (Q3-Q1)
# outlier bawah = Q1 - 1.5 * (Q3-Q1)
```

```python
sns.boxplot(x='sex', y='fare', hue='deck', data=dataku)
```

```python
sns.boxplot(x='sex', y='fare', hue='deck', hue_order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
# Violinplot, menunjukkan distribusi data
# Violinplot kita bisa melihat Q1,Q2,Q3 dan batas outlier seperti di boxplot

sns.violinplot(x='sex', y='fare', palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
sns.violinplot(x='sex', y='fare', hue='deck', hue_order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
sns.violinplot(x='deck', y='fare', order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
# dipisahkan berdasarkan sex

sns.violinplot(x='deck', y='fare', hue='sex', order=['A','B','C','D','E','F','G'], palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
# split bisa digunakan klo hue nya terdiri dari 2 kategori
# dipisahkan berdasarkan sex

sns.violinplot(x='deck', y='fare', hue='sex', order=['A','B','C','D','E','F','G'], palette={'male':'skyblue', 'female':'pink'}, split=True, data=dataku)
```

```python
# split bisa digunakan klo hue nya terdiri dari 2 kategori, split dipisahkan berdasarkan ordernya
# dipisahkan berdasarkan sex

sns.violinplot(x='deck', y='fare', hue='sex', order=['A','B','C','D','E','F','G'], split=False, palette={'male':'skyblue', 'female':'pink'}, data=dataku)
```

```python
sns.stripplot(x='deck', y='fare', order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
# Stripplot tanpa violinplot

sns.stripplot(x='deck', y='fare', hue='sex', dodge=True, palette='Set1', order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
# Stripplot tanpa violinplot

sns.stripplot(x='deck', y='fare', hue='sex', dodge=True, palette='Set1', alpha=0.4, jitter=True, order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
# Tambahkan titik data individual (swarmplot atau stripplot)

# Garis Putih di dalam violin menunjukkan nilai tengah (median) dan kuartil. Bisa mengandung boxplot kecil di dalamnya (jika inner='box' digunakan).
# Titik-titik hitam (stripplot) menampilkan data individual per penumpang. Ini membantu kamu melihat sebaran aslinya secara lebih nyata, bukan hanya dari bentuk violin.
# Sumbu X: deck menunjukkan kategori tempat penumpang tinggal di kapal (A, B, C, dst.).
# Sumbu Y: fare menunjukkan jumlah tarif tiket yang dibayar.

# Sebaran Fare di Deck B sangat luas, ada penumpang yang membayar sangat mahal.
# Deck F, D, G memiliki sebaran sempit dan nilai fare cenderung kecil.
# Perbedaan distribusi antara male dan female terlihat, terutama di beberapa dek seperti C dan B.

sns.violinplot(data=dataku, x='deck', y='fare', hue='sex', split=True, inner=None, palette={'male':'skyblue','female':'pink'}, order=['A','B','C','D','E','F','G'])
sns.stripplot(data=dataku, x='deck', y='fare', hue='sex', dodge=True, palette='dark:.3', alpha=0.4, jitter=True)
```

```python
# Swarmplot

sns.swarmplot(x='deck', y='fare', hue='sex',  order=['A','B','C','D','E','F','G'], data=dataku)
```

```python
sns.violinplot(data=dataku, x='deck', y='fare', hue='sex', split=True, inner=None, palette={'male':'skyblue','female':'pink'}, order=['A','B','C','D','E','F','G'])
sns.swarmplot(x='deck', y='fare', hue='sex',  order=['A','B','C','D','E','F','G'], color='black', data=dataku)
```
