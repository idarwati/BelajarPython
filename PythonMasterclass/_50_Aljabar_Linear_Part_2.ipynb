# Perkalian Vektor dengan Skalar

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math
```

```python
v = np.array([1,2])
sk = [-0.5, 1, 1.5, 3]
```

```python
for i in sk:
  sk_v = i*v
  plt.plot([0, sk_v[0]], [0, sk_v[1]], 'o-', linewidth=3, label='Skalar=%g' %i)

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
v = np.array([1,2])
sk = [-0.5, 1, 1.5]

for i in sk:
  sk_v = i*v
  plt.plot([0, sk_v[0]], [0, sk_v[1]], 'o-', linewidth=3, label='Skalar=%g' %i)

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
v = np.array([1,2])
sk = [-0.5, 1]

for i in sk:
  sk_v = i*v
  plt.plot([0, sk_v[0]], [0, sk_v[1]], 'o-', linewidth=3, label='Skalar=%g' %i)

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

# Penjumlahan Vektor

image.png

```python
v1 = np.array([1,3])
v2 = np.array([3,1])

v3a = v1+v2
v3b = np.add(v1,v2)
```

```python
print(v3a, v3b)
```

```python
plt.plot([0, v1[0]], [0, v1[1]], linewidth=3, label='v1')
plt.plot([0, v2[0]], [0, v2[1]], linewidth=3, label='v2')
plt.plot([0, v3a[0]], [0, v3a[1]], linewidth=3, label='v3a')
plt.axis('square')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot([0, v1[0]], [0, v1[1]], linewidth=3, label='v1')
plt.plot([0, v2[0]]+v1[0], [0, v2[1]]+v1[1], linewidth=3, label='v2')
plt.plot([0, v3a[0]], [0, v3a[1]], linewidth=3, label='v3a')
plt.axis('square')
plt.legend()
plt.grid()
plt.show()
```

```python
p1 = sym.latex(sym.simplify(v1))
p2 = sym.latex(sym.simplify(v2))
p3 = sym.latex(sym.simplify(v3a))

display(Math('%s+%s=%s' %(p1,p2,p3)))
```
