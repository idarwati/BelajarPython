```python
y
```

```python
fig = plt.figure(figsize=(4,4), dpi=200) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(8,4), dpi=100) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(1,4), dpi=100) # figsize digunakan untuk mengubah ukuran grafik (dalam inchi), dpi digunakan agar grafik tidak pecah saat di download
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,8), dpi=100)

axes[0].plot(x,y)
axes[0].set_title('Grafik 1')

axes[1].plot(x,y)
axes[1].set_title('Grafik 2')

plt.tight_layout()

plt.show()
```

# **Menambah Legenda**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Grafik Kuadrat')
ax.plot(x,x**3)

plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')
ax.legend()
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')

ax.legend(loc=(0.2, 0.2)) #untuk menentukan posisi legend, umumnya pilihannya : 'upper left', 'upper right', 'lower left', 'lower right'. Saat ini digunakan posisinya 20% dari kiri dan 20% dari bawah (0.2, 0.2)
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')

ax.legend(loc=(0.5, 0.5)) #untuk menentukan posisi legend, umumnya pilihannya : 'upper left', 'upper right', 'lower left', 'lower right'. Saat ini digunakan posisinya 50% dari kiri dan 50% dari bawah (0.5, 0.5)
plt.show()
```

```python
fig = plt.figure(facecolor='blue', dpi=200) #facecolor digunakan untuk mengubah warna di sekeliling grafik
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')
ax.plot(x,x**4, label='Pangkat 4')

ax.legend(loc=(0.2, 0.2)) #untuk menentukan posisi legend, umumnya pilihannya : 'upper left', 'upper right', 'lower left', 'lower right'. Saat ini digunakan posisinya 20% dari kiri dan 20% dari bawah (0.2, 0.2)
plt.show()
```

```python
fig = plt.figure(facecolor='yellow', dpi=200) #facecolor digunakan untuk mengubah warna di sekeliling grafik
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')
ax.plot(x,x**4, label='Pangkat 4')

ax.grid(True) #untuk menampilkan grid
ax.legend(loc=(0.2, 0.2)) #untuk menentukan posisi legend, umumnya pilihannya : 'upper left', 'upper right', 'lower left', 'lower right'. Saat ini digunakan posisinya 20% dari kiri dan 20% dari bawah (0.2, 0.2)
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='Pangkat 2', lw=4, ls='-')
# linewidth bisa juga dituliskan dengan lw
# linestyle bisa juga dituliskan dengan ls, digunakan untuk mengatur style garis diantaranya {'-','--',':','',(offset, on-off-sew),...}
# marker: marker or style
# markeredgecolor or mec : color
# markeredgewidth or mew : float
# markerfacecolor or mfc : color
# markerfacecoloralt or mfcalt : color
# markersize or ms : float
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.')

ax.grid(True)
ax.legend(loc=(0.2, 0.2))
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='Pangkat 2', lw=1, ls='-', marker='o', markersize=10, mfc='yellow', mew=3, mec='red')
# linewidth bisa juga dituliskan dengan lw
# linestyle bisa juga dituliskan dengan ls, digunakan untuk mengatur style garis diantaranya {'-','--',':','',(offset, on-off-sew),...}
# marker: marker or style
# markeredgecolor or mec : color
# markeredgewidth or mew : float
# markerfacecolor or mfc : color
# markerfacecoloralt or mfcalt : color
# markersize or ms : float
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--', marker='*', markersize=10, mfc='yellow', mew=3, mec='red')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.', marker='+', markersize=10, mfc='yellow', mew=3, mec='red')

ax.grid(True)
ax.legend(loc=(0.2, 0.2))
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='Pangkat 2', lw=4, ls='-', marker='o', markersize=5, mfc='yellow', mew=1, mec='red', mfcalt='black')
# linewidth bisa juga dituliskan dengan lw
# linestyle bisa juga dituliskan dengan ls, digunakan untuk mengatur style garis diantaranya {'-','--',':','',(offset, on-off-sew),...}
# marker: marker or style (bentuk marker)
# markeredgecolor or mec : color (warna marker)
# markeredgewidth or mew : float (ketebalan marker)
# markerfacecolor or mfc : color (warna di dalam marker)
# markerfacecoloralt or mfcalt : color
# markersize or ms : float (ukuran marker)
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--', marker='*', markersize=7, mfc='blue', mew=2, mec='red')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.', marker='+', markersize=10, mfc='yellow', mew=2, mec='red')

ax.grid(True)
ax.legend(loc=(0.2, 0.2))
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='Pangkat 2', lw=4, ls='-', marker='o', markersize=5, mfc='yellow', mew=1, mec='red', mfcalt='black')
# linewidth bisa juga dituliskan dengan lw
# linestyle bisa juga dituliskan dengan ls, digunakan untuk mengatur style garis diantaranya {'-','--',':','',(offset, on-off-sew),...}
# marker: marker or style (bentuk marker)
# markeredgecolor or mec : color (warna marker)
# markeredgewidth or mew : float (ketebalan marker)
# markerfacecolor or mfc : color (warna di dalam marker)
# markerfacecoloralt or mfcalt : color
# markersize or ms : float (ukuran marker)
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--', marker='*', markersize=7, mfc='blue', mew=2, mec='red')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.', marker='+', markersize=10, mfc='yellow', mew=2, mec='red')

ax.grid(True)
ax.legend(loc=(0.2, 0.2))

ax.set_xlim([0,12.5])
ax.set_ylim([0,12.5])

plt.show()
```
