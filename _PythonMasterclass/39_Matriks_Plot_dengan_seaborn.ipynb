Matrix dengan Seaborn

```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
file_id = '1kQHNP1cpFw-44yTaEcbLrvBKY_CtUDwj'
url = f'https://drive.google.com/uc?id={file_id}'
```

```python
dataku = pd.read_csv(url)
```

```python
dataku.head()
```

```python
# Heatmap (Matix Plot) utk data2 numerik

dataku.corr(numeric_only=True)
```

```python
sns.heatmap(dataku.corr(numeric_only=True))
```

```python
sns.heatmap(dataku.corr(numeric_only=True),annot=True)
```

```python
sns.heatmap(dataku.corr(numeric_only=True),annot=True, cmap='Reds')
```

```python
sns.heatmap(dataku.corr(numeric_only=True),annot=True, cmap='Greens')
```

```python
# import data penerbangan

terbang = sns.load_dataset('flights')
```

```python
terbang.head()
```

```python
# membuat heatmap dengan membuat pivot tablenya terlebih dahulu
```

```python
terbang
```

```python
terbang.pivot_table(index='month', columns='year', values='passengers')
```

```python
terbang_pivot = terbang.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(terbang_pivot, cmap='Greens')
```

```python
sns.heatmap(terbang_pivot, cmap='Reds')
```

```python
sns.heatmap(terbang_pivot, cmap='Blues', linecolor='black', linewidths=1)
# linewidth merupakan ukuran ketebalan dari linecolor/batas masing2 heatmap
```

```python
sns.heatmap(terbang_pivot, cmap='Blues', linecolor='white', linewidths=1)
```

```python
sns.heatmap(terbang_pivot, cmap='Reds', linecolor='white', linewidths=1)
```

```python
sns.clustermap(terbang_pivot)
# seaborn melakukan clustering dari data yg telah diimpor
# Klo menggunakan clustermap bulan dan tahunnya jadi tidak berurutan.
# di bagian atas dan kanan grafik juga terdapat diagram pohon clustering (Dendogram)
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm')
# Bulan July and August serta tahun 1959 dan 1960 dikelompokkan karna jumlahnya banyak.
# Tahun 1949 dan 1950 dikelompokkan karna jumlahnya sedikit
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm', standard_scale=1)
# standard_scale digunakan untuk membuat imterpretasi data lebih jelas warnanya
```
